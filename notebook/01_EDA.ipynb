```python
# import some library for Haversine formula.
from math import radians, sin, cos, sqrt, atan2
```

```python
def haversine(lat1, lon1, lat2, lon2):
    R = 6371  # Earth radius in kilometers

    lat1, lon1, lat2, lon2 = map(
        radians,
        [lat1, lon1, lat2, lon2]
    )

    dlat = lat2 - lat1
    dlon = lon2 - lon1

    a = (
        sin(dlat / 2) ** 2
        + cos(lat1) * cos(lat2) * sin(dlon / 2) ** 2
    )

    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    return R * c
```

```python
route_movement["distance_km"] = route_movement.apply(
    lambda x: haversine(
        x["from_latitude"],
        x["from_longitude"],
        x["to_latitude"],
        x["to_longitude"]
    ),
    axis=1
)
```

```python
# calculate speed
route_movement["speed_kmh"] = (
    route_movement["distance_km"] /
    route_movement["time_hours"]
)
```

```python
# spped category
def speed_category(speed):
    if speed > 60:
        return "Fast"
    elif speed >= 30:
        return "Moderate"
    else:
        return "Slow"

route_movement["speed_category"] = route_movement["speed_kmh"].apply(speed_category)
```
