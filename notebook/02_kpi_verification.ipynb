**key business insights as per assignment**

```python
origin_summary = (fact_od.groupby("origin").agg(Container_Count=("cntr_trxn_id","nunique")).sort_values("Container_Count",ascending=False))

print(origin_summary)
```

```python
top_destination = (
    fact_od.groupby("destination_name")
    .agg(Container_Count=("cntr_trxn_id","nunique"))
    .sort_values("Container_Count",ascending=False)
)

print(top_destination.head(10))
```

```python
division_summary = (
    fact_od.groupby("destination_division")
    .agg(Container_Count=("cntr_trxn_id","nunique"))
    .sort_values("Container_Count",ascending=False)
)

print(division_summary)
```

```python
slowest_lane = (fact_od.groupby(
        ["origin", "destination_division"]).agg(
        Avg_Journey_Hours=("journey_hours","mean"),
        Containers=("journey_key","count")).sort_values(
        "Avg_Journey_Hours",
        ascending=False ))

print(slowest_lane.head(10))
```

```python
ncr = ["Delhi Division", "Ambala Division"]

ncr_terminal = (
    fact_od[
        fact_od["destination_division"].isin(ncr)
    ]
    .groupby("origin")
    .agg(
        Containers=("journey_key", "count"),
        Avg_Journey_Time=("journey_hours", "mean")
    )
    .sort_values("Containers", ascending=False)
)

print(ncr_terminal)
```
